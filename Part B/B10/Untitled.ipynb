```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('StudentsPerformance.csv')
print(df.head(5))
```

```python
df.drop(['lunch'],axis=1,inplace=True)
print(df.head(5))
```

```python
df['parental level of education'] = df['parental level of education'].fillna('Not applicable')
print(df.head(5))
```

```python
df['race/ethnicity'] = df['race/ethnicity'].map({
    'group A':'Asian Students',
    'group B':'African Students',
    'group C':'Asian-african Students',
    'group D':'American Students',
    'group E':'European Students'
})
print(df.head(5))
```

```python
ax=sns.countplot(data=df,x='test preparation course',hue='gender',palette='Set1')
ax.set(title='Number of students who took up test preparation course',ylabel='Count',xlabel='Course')
plt.show()
```

```python
ax=sns.countplot(data=df,x='race/ethnicity',hue='gender',palette='YlOrRd')
ax.set(title='Total Number of Male & Female Students belonging to each student group',ylabel='Count',xlabel='Student group')
plt.show()
```

```python
""" No of students who ‘failed’(less than 40), ‘second class’(between 40 & 50). ‘first
class’(between 60 & 75) and ‘distinction’(above 75) in ‘Maths’, ‘Reading’ and
‘Writing’."""

interval=(0,40,50,60,75)
category=['Failed','Second Class','First Class','Distinction']
df['Marks_Cat'] = pd.cut(df['mathscore'],interval,labels=category)
ax=sns.countplot(data=df,x='Marks_Cat',palette='Set1',hue='gender')
ax.set(xlabel='Math Scores',ylabel='Count',title='Math Score of Students')
plt.show()
```

```python
df['Marks_Cat'] = pd.cut(df['reading score'],interval,labels=category)
ax=sns.countplot(data=df,x='Marks_Cat',palette='Set3',hue='gender')
ax.set(xlabel='Reading Scores',ylabel='Count',title='Reading Score of Students')
plt.show()
```

```python
df['Marks_Cat'] = pd.cut(df['writing score'],interval,labels=category)
ax=sns.countplot(data=df,x='Marks_Cat',palette='Set3',hue='gender')
ax.set(xlabel='Writing Scores',ylabel='Count',title='Writing Score of Students')
plt.show()
```
