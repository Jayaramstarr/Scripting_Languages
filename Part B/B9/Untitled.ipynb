```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
print(df.head(5))
print(df.info())
print(df.describe())
```

```python
df.drop(['Sepal_Length'],axis=1,inplace=True)
print(df.head(5))
```

```python
df.groupby(['Class']).mean()['Petal_Width']
```

```python
plt.figure(figsize=[12,6])
ax=sns.countplot(x='Sepal_Width',hue='Class',palette='YlOrRd',data=df)
ax.set(title='Number of flowers of each species existing for each value of sepal width',ylabel='Count',xlabel='Sepal Width')
plt.show()
plt.tight_layout()
```

```python
interval = (0,1,2,4)
category = ['<1','1 to 2','>2']
df['Petal_Catg'] = pd.cut(df['Petal_Width'],interval,labels=category)
ax = sns.countplot(data = df,x = 'Petal_Catg',hue='Class',palette='YlOrRd')
ax.set(title='Petal Width',xlabel='Category of Petals',ylabel='No. of flowers')
plt.show()
```

```python
plt.figure(figsize=[12,6])
ax=sns.countplot(data=df[df['Class']=='Iris-versicolor'],palette='Set1',x='Sepal_Width')
ax.set(title='Number of Iris-Versicolour species according to the value of Sepal Width',xlabel='Sepal Width',ylabel='Count')
plt.show()
```

```python
plt.figure(figsize=[12,6])
ax=sns.countplot(data=df[df['Class']=='Iris-virginica'],palette='Set1',x='Sepal_Width')
ax.set(title='Number of Iris-Virginica species according to the value of Sepal Width',xlabel='Sepal Width',ylabel='Count')
plt.show()
```
