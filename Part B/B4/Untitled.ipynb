```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_titanic=pd.read_csv('train.csv')

df_titanic['Survived'] = df_titanic['Survived'].map({
    1:'Survived',
    0:'Dead'
})
df_titanic.drop(['Name','Parch','Ticket','Fare','PassengerId','SibSp'],axis=1,inplace=True)
df_titanic['Pclass'] = df_titanic['Pclass'].map({
    1:'Luxury Class',
    2:'Economy Class',
    3:'Lower Class'
})
df_titanic['Embarked'] = df_titanic['Embarked'].map({
    'S':'Southampton',
    'C':'Cherbourg',
    'Q':'Queenstown'
})
df_titanic['Cabin'] = df_titanic['Cabin'].fillna('S')
print(df_titanic.head(5))


ax = sns.countplot(x='Pclass',hue='Survived',palette='Set1', data=df_titanic)
ax.set(title='Survival rate based on Passenger class',xlabel='Passenger Class',ylabel='Count')
plt.show()

print(pd.crosstab(df_titanic['Survived'],df_titanic['Sex']))
ax = sns.countplot(data=df_titanic,x='Sex',hue='Survived',palette='Set2')
ax.set(title='Survival rate based on sex', xlabel='Sex', ylabel='Count')
plt.show()

intervals=[0,18,35,60,120]
categories=['Children','Teens','Adult','Old']
df_titanic['Age_cats'] = pd.cut(df_titanic['Age'],intervals, labels = categories)
ax = sns.countplot(data=df_titanic,x='Age_cats',hue='Survived',palette='Set3')
ax.set(title='Survival rate based on age', xlabel='Ages', ylabel='Count')
plt.show()

ax = sns.countplot(data=df_titanic,x='Embarked',hue='Survived',palette='Set1')
ax.set(title='Survival rate based on embarked place', xlabel='Emabarked places', ylabel='Count')
plt.show()
```
